```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('main_theme')
```

```python
men_df = pd.read_csv('/Users/haroontaylor/Documents/Sports_stats/Docs/CSVs/Grand_slam_winners_men.csv')
women_df = pd.read_csv('/Users/haroontaylor/Documents/Sports_stats/Docs/CSVs/Grand_slam_winners_women.csv')
```

```python
def summarise_df(df, threshold):
    df = (df
          .query('Year >= {}'.format(threshold)).iloc[:, 1:5]
          .apply(lambda x: x.str.split('(').str[0])
          .apply(pd.Series.value_counts))
    df['Total'] = df.sum(axis=1)
    df = (df
          .sort_values(by='Total', ascending=False)
          .drop('Total', axis=1))
    
    return df
```

```python
men_2006 = summarise_df(men_df, 2006)
women_2006 = summarise_df(women_df, 2006)
```

```python
print('Male Grand Slam winners since 2006:', len(men_2006))
men_2006.plot(kind='bar', stacked=True, color=['#0297DB', 'C1', 'C2', '#3C638E'], figsize=(20, 10))
plt.title('Male Grand Slam winners since 2006')
plt.ylim((0, 18))
plt.xlabel('Player')
plt.ylabel('Titles')
plt.legend(fontsize=20)
plt.show()
```

```python
print('Female Grand Slam winners since 2006:', len(women_2006))
women_2006.plot(kind='bar', stacked=True, color=['#0297DB', 'C1', 'C2', '#3C638E'], figsize=(20, 10))
plt.title('Female Grand Slam winners since 2006')
plt.xlabel('Player')
plt.ylabel('Titles')
plt.legend(fontsize=20)
plt.show()
```
